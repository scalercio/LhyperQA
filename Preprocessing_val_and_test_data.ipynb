```python
import pandas as pd
data = pd.read_csv('Elementary-NDMC-Dev.csv')
print(len(data))
data["question"][0]
```

```python
data["aux"]=data["question"].apply(lambda x: x.replace(" (A) ","scalercio").replace(" (B) ","scalercio").replace(" (C) ","scalercio").replace(" (D) ","scalercio").split("scalercio"))
data["aux"][122]
```

```python
df=data.loc[data["AnswerKey"]!="E"]
df.index = range(len(df))
print(len(df))
df
```

```python
df["q1"]=df["aux"].apply(lambda x: x[0])
df["a"]=df["aux"].apply(lambda x: x[1])
df["b"]=df["aux"].apply(lambda x: x[2])
df["c"]=df["aux"].apply(lambda x: x[3])
```

```python
import numpy as np
df['d']=np.nan
for index, row in df.iterrows():
    #print(len(row['aux']))
    if(len(row['aux'])>=5):
        df['d'][index]=df["aux"][index][4]
    else:
        df['d'][index]="."
```

```python
import re
# Preprocess the reviews
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"([.,!?])", r" \1 ", text)
    text = re.sub(r"[^a-zA-Z.,!?]+", r" ", text)
    return text
    
df.q1 = df.q1.apply(preprocess_text)
df.a = df.a.apply(preprocess_text)
df.b = df.b.apply(preprocess_text)
df.c = df.c.apply(preprocess_text)
df.d = df.d.apply(preprocess_text)
```

```python
for index, row in df.iterrows():
    if(row["q1"].split(" ")[-1]==""):
        df["q1"][index]= row["q1"].split(" ")[:-1]
    else:
        df["q1"][index]= row["q1"].split(" ")
```

```python
for index, row in df.iterrows():
    if(row["a"].split(" ")[-1]==""):
        df["a"][index]= row["a"].split(" ")[:-1]
    else:
        df["a"][index]= row["a"].split(" ")
        
    if(row["b"].split(" ")[-1]==""):
        df["b"][index]= row["b"].split(" ")[:-1]
    else:
        df["b"][index]= row["b"].split(" ")
    
    if(row["c"].split(" ")[-1]==""):
        df["c"][index]= row["c"].split(" ")[:-1]
    else:
        df["c"][index]= row["c"].split(" ")
        
    if(row["d"].split(" ")[-1]==""):
        df["d"][index]= row["d"].split(" ")[:-1]
    else:
        df["d"][index]= row["d"].split(" ")
    
```

```python
df["len_q1"]=df["q1"].apply(lambda x: len(x))
df["len_a"]=df["a"].apply(lambda x: len(x))
df["len_b"]=df["b"].apply(lambda x: len(x))
df["len_c"]=df["c"].apply(lambda x: len(x))
df["len_d"]=df["d"].apply(lambda x: len(x))
```

```python
df["a1"]=np.nan
df["wa1"]=np.nan
df["wa2"]=np.nan
df["wa3"]=np.nan
for index, row in df.iterrows():
    if(row['AnswerKey']=="A"):
        df["a1"][index] = df["a"][index]
        df["wa1"][index] = df["b"][index]
        df["wa2"][index] = df["c"][index]
        df["wa3"][index] = df["d"][index]
    elif (row['AnswerKey']=="B"):
        df["a1"][index] = df["b"][index]
        df["wa1"][index] = df["a"][index]
        df["wa2"][index] = df["c"][index]
        df["wa3"][index] = df["d"][index]
    elif (row['AnswerKey']=="C"):
        df["a1"][index] = df["c"][index]
        df["wa1"][index] = df["b"][index]
        df["wa2"][index] = df["a"][index]
        df["wa3"][index] = df["d"][index]
    elif (row['AnswerKey']=="D"):
        df["a1"][index] = df["d"][index]
        df["wa1"][index] = df["b"][index]
        df["wa2"][index] = df["c"][index]
        df["wa3"][index] = df["a"][index]
```

```python
df["len_a1"]=df["a1"].apply(lambda x: len(x))
df["len_wa1"]=df["wa1"].apply(lambda x: len(x))
df["len_wa2"]=df["wa2"].apply(lambda x: len(x))
df["len_wa3"]=df["wa3"].apply(lambda x: len(x))
```

```python
df1=df[["AnswerKey","question","q1","a1","wa1", "wa2", "wa3", "len_a1", "len_wa1", "len_wa2", "len_wa3","len_q1","a","b", "c", "d","len_a", "len_b", "len_c", "len_d"]]
df1.to_csv('/home/arthur/learning/ml_and_dataScience/LhyperQA/validation_data.csv',index=False)
```
